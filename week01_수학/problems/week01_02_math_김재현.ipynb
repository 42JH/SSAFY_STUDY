# 1. 베르트랑 공준

```python
n = int(input())

while n: # 0이 입력되기 전까지 반복
    prime = [True] * (2*n + 1)
    prime[0] = False
    prime[1] = False

    for i in range(2, int((2*n)**0.5)+1):
        if prime[i]:
            for j in range(i * i, (2*n) + 1, i):
                prime[j] = False
    
    # 2n 크기의 에라토스테네스의 체 형성

    print(prime[n+1:2*n+1].count(True))
    # n보다 크고 2n보다 작거나 같은 구간을 탐색하여 소수의 개수 출력
    
    n = int(input())
```

# 2. 골드바흐 파티션

```python
MAX = 1000001

def sieve(n):  # 에라토스테네스의 체
    prime = [True] * (n + 1)
    prime[0] = False
    prime[1] = False

    for i in range(2, int(n**0.5) + 1):
        if prime[i]:
            for j in range(i * i, n+1, i):
                prime[j] = False
    
    return prime

prime = sieve(MAX)

T = int(input())
for test_case in range(T):
    N = int(input())
    cnt = 0

    for a in range(2, N//2 + 1):
        if prime[a] and prime[N-a]: #(2,N-2), (3,N-3), ... 순으로 탐색
            cnt += 1  # 만약 두 수가 모두 소수이면 cnt값 1 증가

    print(cnt)
```

# 3. 숨바꼭질 6

```python
def gcd(a,b):  # 유클리드 호제법
    while b:
        a, b = b, a % b
    return a

N, S = map(int, input().split())
A = list(map(int, input().split()))
distance = [abs(S-x) for x in A] # 수빈이의 위치와 동생들의 거리를 리스트안에 넣음

result = distance[0]
for i in distance[1:]: # 유클리드 호제법을 여러번 사용
    result = gcd(result, i) # 결과값을 result에 할당
    if result == 1: 
        break

print(result)

```

# 4. 백대열 (14490번)

```python
# 두 수가 주어지는 경우 최대한 약분하여 n:m 형식으로 출력하는 문제.
# 두 수의 최대공약수를 구한 후 두 수를 약분하여 형식에 맞게 출력한다.

def gcd(a,b):
    while b:
        a, b = b, a % b
    return a

n, m = map(int, input().split(':'))

x = gcd(n,m)
print(f'{n//x}:{m//x}')
```

# 5. 최대공약수 (1850번)

```python
# 모든 자리가 1로만 이루어져있는 두 자연수가 주어진다.
# 이때 이 자연수들의 최대공약수를 구하는 프로그램을 작성하시오.
# 입력하는 A,B는 자연수들의 자릿수이다. ex) A = 3 -> 111

def gcd(x,y):
    while y:
        x, y = y, x % y
    return x


A, B = map(int, input().split())

result = gcd(A,B)

print((10**result)//9)
```
