## 1

```python
from collections import deque

def solve():
    for _ in range(10): # 10개 테스트 케이스 루프
        tc_num = input()
        space = [list(map(int, input())) for _ in range(16)]
        visited = [[0] * 16 for _ in range(16)]
        
        queue = deque()

        # 출발점 찾기
        start_r, start_c = 1, 1 
        for i in range(16):
            for j in range(16):
                if space[i][j] == 2:
                    start_r, start_c = i, j
        
        queue.append([start_r, start_c])
        visited[start_r][start_c] = 1
        ans = 0

        def bfs():
            global ans 
            while queue:
                cur_r, cur_c = queue.popleft()

                # 현재 위치가 목적지인지 확인
                if space[cur_r][cur_c] == 3:
                    ans = 1
                    return # 찾았으면 즉시 종료

                # 4방향 탐색
                for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    # 기준 좌표(cur_r, cur_c)에서 이동해야 함
                    nr, nc = cur_r + dr, cur_c + dc

                    if 0 <= nr < 16 and 0 <= nc < 16:
                        if visited[nr][nc] == 0 and space[nr][nc] != 1:
                            queue.append([nr, nc])
                            visited[nr][nc] = 1
        
        bfs() # 정의한 함수
        print(f"#{tc_num} {ans}")

# 실행
solve()
```

## 2

```python
T = int(input())

for test_case in range(T):

    arr, n = map(int, input().split())
    arr = list(str(arr))

    visited = set()

    ans = 0

    def dfs(cnt):
        global ans

        # 중복 체크.
        cur_str = ''.join(map(str,arr))
        if (cur_str, cnt) in visited:
            return
        
        visited.add((cur_str, cnt))

        # n을 다 쓴경우
        if cnt == n:
            ans = max(ans, int(cur_str))
            return

        # 모든 가능한 두자리를 선택하여 교환..
        length = len(arr)
        for i in range(length):
            for j in range(i + 1, length):
                arr[i], arr[j] = arr[j], arr[i] # 교체
                dfs(cnt + 1)
                arr[i], arr[j] = arr[j], arr[i] # 원상복구

    dfs(0)
    print(f'#{test_case + 1} {ans}')
```
