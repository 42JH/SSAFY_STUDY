# 1. 베르트랑 공준

```python
def sol1(n):
    is_prime = [True] * (n + 1)
    is_prime[0] = False
    if n >= 1:
        is_prime[1] = False
    
    p = 2
    while p * p <= n:
        if is_prime[p]:
            for multiple in range(p * p, n + 1, p):
                is_prime[multiple] = False
        p += 1
    return sum(is_prime)

def sol2(n):
    return sol1(2*n) - sol1(n)
    
    
n = int(input())
print(sol2(n))
```
