# 1. 베르트랑 공준

```python
# 임의의 자연수 n 에 대해 2n - n 사이 소수의 갯수를 구하는 문제.
# 따라서 2n 의 소수의 갯수 - n 소수의 갯수로 원하는 값을 찾을 수 있다.

def sol1(n): 
    is_prime = [True] * (n + 1) 
    is_prime[0] = False
    if n >= 1:
        is_prime[1] = False
    
    p = 2
    while p * p <= n:
        if is_prime[p]:
            for multiple in range(p * p, n + 1, p):
                is_prime[multiple] = False
        p += 1
    return sum(is_prime)

def sol2(n):
    return sol1(2*n) - sol1(n)
n = 1
while n != 0:
    n = int(input())
    print(sol2(n))
```

# 2. 골드바흐 파티션

```python
def sol1(n): # 골드바흐의 체. 주어진 n 중 소수 판별 함수.
    prime = [True] * (n + 1)
    prime[0] = False
    prime[1] = False

    for i in range(2, int(n**0.5) + 1):
        if prime[i]:
            for j in range(i * i, n + 1, i):
                prime[j] = False
    return prime

MAX = 100000 
prime = sol1(MAX) # 미리 최대값 100,000 의 소수 배열을 계산

def sol2(n): # 각 n 을 만들수 잇는 소수의 짝 계산
    c = 0
    for i in range(2, n//2 + 1): # 2부터 n//2 까지 순회
        if prime[i] and prime[n - i]: # 만약 i 번째, n-1 번째 숫자가 소수인 경우.
            c += 1 # 가짓수 + 1
    return c


T = int(input()) 
for _ in range(T):
    n = int(input())
    print(sol2(n))
```

# 3. 숨바꼭질 6

```python
import math
# 주어진 거리 들을 모두 갈수있는 x 를 구하는 문제는 최대 공약수를 찾는 문제와 같음
# 따라서 위치 s 와 위치 a 의 거리를 구한 후 그 거리들로 찾을수 있다.

x = list(map(int, input().split()))
N, S = x[0], x[1]
A = list(map(int, input().split()))

ans = abs(A[0] - S)
for i in range(1, len(A)):
    dist = abs(A[i] - S)
    ans = math.gcd(ans, dist)

print(ans)
```
