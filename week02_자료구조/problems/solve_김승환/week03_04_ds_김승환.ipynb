## 1.[Silver IV] 큐 - 10845

```python
from collections import deque

n_input = input() 
if n_input: # n이 안들어올때까지..
    n = int(n_input)
    queue = deque()
    for _ in range(n):
        command = input().split()
        if not command: continue
        
        cmd_type = command[0]
        if cmd_type == "push":
            queue.append(command[1])
        elif cmd_type == "pop":
            print(queue.popleft() if queue else -1)
        elif cmd_type == "size":
            print(len(queue))
        elif cmd_type == "empty":
            print(0 if queue else 1)
        elif cmd_type == "front":
            # popleft()를 하면 데이터가 사라지므로 인덱스 [0] 사용
            print(queue[0] if queue else -1)
        elif cmd_type == "back":
            # pop()을 하면 데이터가 사라지므로 인덱스 [-1] 사용
            print(queue[-1] if queue else -1)
```

## 2. [Silver II] 최대 힙 - 11279

```python
import heapq
import sys

n_input = input()
if n_input:
    n = int(n_input)
    max_heap = []

    for _ in range(n):
        try:
            x = int(input())
            if x == 0:
                # 최대 힙 구현을 위해 음수로 넣었으므로 다시 양수로 출력
                print(-heapq.heappop(max_heap) if max_heap else 0)
            else:
                # Python의 heapq는 최소 힙이므로 음수로 변환해 삽입
                heapq.heappush(max_heap, -x)
        except EOFError:
            break
```

## 3. [Silver IV] 덱 - 10866

```python
from collections import deque
import sys

n_input = input()
if n_input:
    n = int(n_input)
    dq = deque()

    for _ in range(n):
        command = input().split()
        if not command: continue
        
        cmd_type = command[0]

        if cmd_type == "push_front":
            dq.appendleft(command[1])
        elif cmd_type == "push_back":
            dq.append(command[1])
        elif cmd_type == "pop_front":
            print(dq.popleft() if dq else -1)
        elif cmd_type == "pop_back":
            print(dq.pop() if dq else -1)
        elif cmd_type == "size":
            print(len(dq))
        elif cmd_type == "empty":
            print(0 if dq else 1)
        elif cmd_type == "front":
            print(dq[0] if dq else -1)
        elif cmd_type == "back":
            print(dq[-1] if dq else -1)
```
